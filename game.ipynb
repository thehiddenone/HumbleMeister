```python
import humblemeister as hmt

from chess import BLACK, WHITE
from time import sleep

model = hmt.ChessModel.load("env/out/model_giant_04072026_2000_b3")
```

```python
game = hmt.ChessGame(model, 5.0)
game.start_game(WHITE)
```

```python
from IPython.display import clear_output, display

while not game.board.is_game_over():
    game.apply_move(game.sample_move())
    clear_output(wait=True)
    display(game.render())
```
